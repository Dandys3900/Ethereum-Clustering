```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV file
data = pd.read_csv("1procento.csv", usecols=["timestamp", "/.used"])
data["timestamp"] = pd.to_datetime(data["timestamp"]).dt.strftime("%H:%M:%S")

# Set seaborn style
#sns.set_theme(style="whitegrid")

# Create the plot
plt.figure(figsize=(20, 5))
sns.lineplot(x=data["timestamp"], y=data["/.used"], color="b", label="Data Trend")

# Titles and labels
plt.ylabel("GB")
plt.title("Linear Chart from CSV Data")
plt.legend()
plt.gca().xaxis.set_visible(False)

# Show plot
plt.show()

# Load CSV file
data = pd.read_csv("1procento.csv", usecols=["timestamp", "br-d9dc76c093c8.bytes_recv_rate_per_sec", "br-d9dc76c093c8.bytes_sent_rate_per_sec"])
data["timestamp"] = pd.to_datetime(data["timestamp"]).dt.strftime("%H:%M:%S")

# Set seaborn style
#sns.set_theme(style="whitegrid")

sentdata = data["br-d9dc76c093c8.bytes_sent_rate_per_sec"]
recvdata = data["br-d9dc76c093c8.bytes_recv_rate_per_sec"]

# Create the plot
plt.figure(figsize=(20, 5))
sns.lineplot(data=sentdata, color="b", label="Sent/s")
sns.lineplot(data=recvdata, color="r", label="Recv/s")

# Titles and labels
plt.ylabel("B")
plt.title("Linear Chart from CSV Data")
plt.legend()
plt.gca().xaxis.set_visible(False)

# Show plot
plt.show()

# Load CSV file
data = pd.read_csv("1procento.csv", usecols=["timestamp", "br-d9dc76c093c8.bytes_recv", "br-d9dc76c093c8.bytes_sent"])
data["timestamp"] = pd.to_datetime(data["timestamp"]).dt.strftime("%H:%M:%S")

# Set seaborn style
#sns.set_theme(style="whitegrid")

sentdata = data["br-d9dc76c093c8.bytes_sent"]
recvdata = data["br-d9dc76c093c8.bytes_recv"]

# Create the plot
plt.figure(figsize=(20, 5))
sns.lineplot(data=sentdata, color="b", label="Sent")
sns.lineplot(data=recvdata, color="r", label="Recv")

# Titles and labels
plt.ylabel("B")
plt.title("Linear Chart from CSV Data")
plt.legend()
plt.gca().xaxis.set_visible(False)

# Show plot
plt.show()
```
